# Exerc√≠cios da Aula 04

```python
import pandas as pd
import io
import numpy as np
import folium
import plotly.express as px

from google.colab import files
```

```python
arq = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(arq['AB_NYC_2019.csv']))
```

```python
df1 = df.dropna()
```

```python
df1.head()
```

### **1. Desenhe um mapa da cidade de Nova York, localizando somente os top 10 im√≥veis com mais avalia√ß√£oes, por regi√£o**

```python
cols = ['number_of_reviews', 'neighbourhood_group', 'latitude', 'longitude']
data_plot = df1.loc[:, cols].groupby('neighbourhood_group').head(10)
data_plot = data_plot.sort_values(['neighbourhood_group', 'number_of_reviews'], ascending=[True, False]).reset_index()

map = folium.Map()

for index, location_info in data_plot.iterrows():
  folium.Marker([location_info['latitude'], location_info['longitude']], popup=location_info['neighbourhood_group']).add_to(map)

map
```

### **2. Desenhe um map da cidade de Nova York, localizando somente os top 50 maiores pre√ßos**

```python
cols = ['name', 'price', 'latitude', 'longitude']
data_plot = df1.loc[:, cols].sort_values(['price'], ascending=[False]).head(50)

map = folium.Map()

for index, location_info in data_plot.iterrows():
  folium.Marker([location_info['latitude'], location_info['longitude']], popup=location_info['name']).add_to(map)

map
```

### **3. Desenhe um map da cidade de Nova York, localizando somente os top 10 menores pre√ßos, por regi√£o**

```python
cols = ['number_of_reviews', 'neighbourhood_group', 'latitude', 'longitude']
data_plot = df1.loc[:, cols].groupby('neighbourhood_group').tail(10)
data_plot = data_plot.sort_values(['neighbourhood_group', 'number_of_reviews'], ascending=[True, False]).reset_index()

map = folium.Map()

for index, location_info in data_plot.iterrows():
  folium.Marker([location_info['latitude'], location_info['longitude']], popup=location_info['neighbourhood_group']).add_to(map)

map
```

### **4. Desenhe um gr√°fico de barras, mostrando a m√©dia de aluguel di√°ria de cada regi√£o**

```python
aluguel_diario_regiao = df1.groupby('neighbourhood_group')['price'].mean().reset_index()
aluguel_diario_regiao
px.bar(aluguel_diario_regiao, x='neighbourhood_group', y='price')
```

### **5. Desenhe um gr√°fico de barras, mostrando a mediana de aluguel di√°ria de cada regi√£o**

```python
aluguel_diario_regiao = df1.groupby('neighbourhood_group')['price'].median().reset_index()
aluguel_diario_regiao
px.bar(aluguel_diario_regiao, x='neighbourhood_group', y='price')
```

### **6. Desenhe um gr√°fico de barras, mostrando a soma de todos os valores da diaria de cada regi√£o**

```python
aluguel_diario_regiao = df1.groupby('neighbourhood_group')['price'].sum().reset_index()
aluguel_diario_regiao
px.bar(aluguel_diario_regiao, x='neighbourhood_group', y='price')
```

### **7. Desenhe um gr√°fico de barras, mostrando o menor valor de di√°ria de cada regi√£o**

```python
aluguel_diario_regiao = df1.groupby('neighbourhood_group')['price'].min().reset_index()
aluguel_diario_regiao
px.bar(aluguel_diario_regiao, x='neighbourhood_group', y='price')
```
