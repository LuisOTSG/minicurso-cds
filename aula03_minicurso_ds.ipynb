## **Como é a variação dos preços dos imóveis em NY?**

```python
des_pad = np.std(df1['price'])
print(f"O desvio padrão dos preços é R${des_pad:.2f}")

price_avg = np.mean(df1['price'])
print(f"A média de preços é R${price_avg:.2f}")
```
